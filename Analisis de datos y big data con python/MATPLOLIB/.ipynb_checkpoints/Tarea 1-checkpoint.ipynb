```python
import matplotlib.pyplot as plt
import numpy as np
```

#### EJERCICIO 1

##### Crea un gráfico de barras que compare las ventas trimestrales de tresproductos.
##### Incluye título, etiquetas y colores personalizados.

```python
#VALORES PARA LA TABLA
Productos = ["Producto_1","Producto_2","Producto_3"]
Valores =[150,200,50]

#TIPO DE GRÁFICA, VALORES USADOS Y COLOR DE RELLENO
plt.bar(Productos, Valores, color='red')
#TITULO DEL GRAFICO
plt.title("Ventas trimestrales")
#MOSTRAR EL GRÁFICO
plt.show()
```

#### EJERCICIO 2

##### Usando datos aleatorios, crea un histograma que muestre la distribución de edades de 200 personas.
##### Ajusta el número de bins para obtener una visualización clara.

```python
datos=[(np.random.randint(3, 90,size=200))]
plt.hist(datos,bins=50,color='blue',edgecolor='black')
plt.title("Distribución de edades")
plt.xlabel("Edades")
plt.ylabel("Frecuencia")
plt.show()
```

#### EJERCICIO 3

##### Representa con un gráfico de líneas la evolución del precio medio de la gasolina durante 12 meses.
##### Añade una anotación donde se alcance el precio máximo.

```python
precio = [12.30,
          10.50,
          50.10,
          5.60,
         15.45,
         7.12,
         8.15,
         6.15,
         20.50,
         80.20,
         90.00,
         75.89]
meses = ["Ene","Febr","Marz","Abr","May","Jun","Jul","Ago","Sept","Oct","Nov","Dic"]
plt.plot(meses,precio,color='green',linestyle='-')
plt.title("Media de precio anual")
plt.xlabel("Meses del año")
plt.ylabel("euros (€)")
plt.show()
```
