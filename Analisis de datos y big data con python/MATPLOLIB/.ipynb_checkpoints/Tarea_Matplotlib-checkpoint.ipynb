#### EJERCICIO 4

##### Combina varios subgráficos (subplots) en una figura con:
##### Un gráfico de barras,
##### Un gráfico de líneas
##### Un histograma
##### Analiza qué información aporta cada uno

```python
fig, axs = plt.subplots(1,3, figsize=(15,4))

axs[0].bar(["Producto_1","Producto_2","Producto_3"], [150,200,50], color='orange')
axs[0].set_title("Barras")
axs[0].set_xlabel("Productos")
axs[0].set_ylabel("Valor €")


meses = ["Ene","Febr","Marz","Abr","May","Jun","Jul","Ago","Sept","Oct","Nov","Dic"]
precio = [12.30,10.50,50.10,5.60,15.45,7.12,8.15,6.15,20.50,80.20,90.00,75.89]

axs[1].plot(meses,precio,color='green',linestyle='-')
axs[1].set_title("Grafica Lineas")
axs[1].set_xlabel("Meses")
axs[1].set_ylabel("Valor €")

datos=[(np.random.randint(3, 90,size=200))]

axs[2].hist(datos,bins=50,color='blue',edgecolor='black')
axs[2].set_title("Distribución de edades")
axs[2].set_xlabel("Edades")
axs[2].set_ylabel("Frecuencia")


plt.show()
```

##### El de barras aporta la diferencia de precio entre los diferentes productos que hay
##### El de líneas aporta la evolución del precio con respecto al tiempo en este caso cada mes
##### El de edades aporta la cantidad de veces aproximadamente que aparecen las edades que se encuentran dentro de los datos
