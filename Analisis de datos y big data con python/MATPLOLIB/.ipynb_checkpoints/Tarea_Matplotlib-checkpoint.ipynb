### BLOQUE 3 - Visualización de resultados Pandas

#### EJERCICIO 8 - Ventas (Ej.6 de Pandas - Tarea3/UD1):

```python
#1. Carga el archivo ventas.csv con read_csv().
df = pd.read_csv("ventas.csv")
impor_med_catg = df.groupby('categoria')['importe'].mean()
categoria = impor_med_catg.index #saco las cabeceras de cada uno de los productos que hace la media
```

##### Genera un gráfico de barras con el importe medio por producto o categoría.

```python
plt.figure(figsize=(12,6))
plt.bar(categoria, impor_med_catg, color='skyblue')
plt.title('Medio de Venta por Categoría',fontsize=16)
plt.xlabel('Categoría de producto',fontsize=12)
plt.ylabel('Importe medio €',fontsize=12)
plt.grid(axis='y',linestyle='--')
plt.tight_layout()
plt.show()
```

##### Añade etiquetas de valor y una línea horizontal indicando la media general.

```python
media_general= df['importe'].mean()
plt.figure(figsize=(12,6))
barras= plt.bar(categoria, impor_med_catg, color='skyblue')
# Recorre cada barra y añade un texto en la parte superior
for bar in barras:
    yval = bar.get_height()
    # Usamos f-string para formatear el valor a 2 decimales
    plt.text(
        bar.get_x() + bar.get_width()/2.0,#Posición X (centro de la barra)
        yval,#Posición Y (altura de la barra)
        f'{yval:.2f}',#Texto de la etiqueta
        ha='center',#horizontal
        va='bottom'#vertical
    )

# B. AÑADIR LÍNEA HORIZONTAL DE LA MEDIA GENERAL
plt.axhline(
    media_general,# Posición Y de la línea
    color='red',# Color de la línea
    linestyle='--',# Tipo de línea
    linewidth=1.5,# Grosor de la línea
)
plt.title('Medio de Venta por Categoría',fontsize=16)
plt.xlabel('Categoría de producto',fontsize=12)
plt.ylabel('Importe medio €',fontsize=12)
plt.grid(axis='y',linestyle='--')
plt.tight_layout()
plt.show()
```

#### EJERCICIO 9 - Análisis musical (Ej.10 de Pandas - Tarea3/UD1):

```python
#1. Carga el archivo tecno.csv con read_csv().
df = pd.read_csv("tecno.csv")

conteo_generos = df['genero'].value_counts()
labels = conteo_generos.index
sizes = conteo_generos.values

usuarios_activos = df['usuario'].value_counts().head(10).sort_values(ascending=True)
user_labels = usuarios_activos.index
user_counts = usuarios_activos.values
#print(df['usuario'].value_counts().head(10))
```

##### Representa un gráfico circular (pie) con la proporción de géneros musicales.

```python
plt.figure(figsize=(10, 10))
plt.pie(
    sizes,
    labels=labels,
    autopct='%1.1f%%',# Formato para mostrar el porcentaje dentro de cada porción
    startangle=90,# Comienza el primer segmento en la parte superior
    wedgeprops={'edgecolor': 'black'}# Añade un borde para mayor claridad
)

# Asegura que el gráfico sea un círculo perfecto
plt.axis('equal')

plt.title('Proporción de Géneros Musicales', fontsize=16)
plt.tight_layout()
plt.show()
```

##### Crea además un gráfico de barras horizontales con los usuarios más activos.

```python
plt.subplot(1, 2, 2) # Define el segundo subplot (1 fila, 2 columnas, posición 2)
plt.barh(user_labels, user_counts, color='darkorange')

# Títulos y Etiquetas
plt.title('Top 10 Usuarios Más Activos', fontsize=14)
plt.xlabel('Número de Actividades/Registros', fontsize=12)
plt.ylabel('Usuario', fontsize=12)
plt.grid(axis='x', linestyle='--')
plt.tight_layout() # Ajusta el diseño para evitar solapamientos entre subplots

plt.show()
```

##### Pregunta abierta: ¿Qué gráfico comunica mejor el patrón de consumo musical?

###### El gráfico circular da una clarra distribución segun los gustos musicales, y el grafico de barras horizontales en este caso todos los valores en el que se representan son los mismos pero para ver una mejor visualización de la frecuencia es muy bueno ya que se representan las barras mas largas a simple vista como una frecuencia mayor
