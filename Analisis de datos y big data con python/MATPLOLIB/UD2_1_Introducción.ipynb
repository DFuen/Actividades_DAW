# 1.3 Tipos de datos y cómo representarlos visualmente

## 1.3.1.- Variable categórica

```python
import matplotlib.pyplot as plt
```

```python
paises = ['España', 'Francia', 'Italia', 'Alemania']
personas = [45, 38, 60, 83]
```

```python
plt.bar(paises, personas, color='skyblue')
plt.title("Población por país (en millones)")
plt.xlabel("País")
plt.ylabel("Millones de habitantes")
plt.show()
```

## 1.3.2.- Variable numérica

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Generamos 100 edades simuladas (entre 18 y 65 años)
edades = np.random.randint(18, 66, 100)
```

```python
plt.hist(edades, bins=10, color='lightgreen', edgecolor='black')
plt.title("Distribución de edades")
plt.xlabel("Edad")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# HISTOGRAMA
```

```python
import seaborn as sns
sns.boxplot(x=edades, color='orange')
plt.title("Diagrama de caja de edades")
plt.show()
```

## 1.3.3.- Variable temporal

```python
import matplotlib.pyplot as plt
```

```python
meses = ['Ene','Feb','Mar','Abr','May','Jun']
ventas = [120, 150, 170, 160, 190, 210]
```

```python
plt.plot(meses, ventas, marker='o', color='teal')
plt.title("Evolución de ventas (enero–junio)")
plt.xlabel("Mes")
plt.ylabel("Ventas (€)")
plt.grid(True)
plt.show()
```

# 1.7 Errores comunes al visualizar datos

## 1. Usar gráficos de sectores para demasiadas categorías

```python
import matplotlib.pyplot as plt

categorias = ['A','B','C','D','E','F','G','H']
valores = [15, 10, 12, 9, 8, 7, 6, 5]

plt.pie(valores, labels=categorias)
plt.title("Demasiadas categorías en un gráfico de sectores")
plt.show()
```

### Solución:Agrupar las categorías pequeñas en una sola

```python
# Agrupar las categorías pequeñas en una sola
categorias_simpl = ['A','B','C','Otras']
valores_simpl = [15, 10, 12, sum(valores[3:])]

plt.pie(valores_simpl, labels=categorias_simpl, autopct='%1.1f%%')
plt.title("Gráfico de sectores simplificado")
plt.show()
```

## 3. Emplear colores similares para categorías distintas

```python
categorias = ['A','B','C']
valores = [30, 25, 20]

plt.bar(categorias, valores, color=['lightblue','skyblue','deepskyblue'])
plt.title("Colores demasiado similares (dificultan la lectura)")
plt.show()
```

### Solución

```python
plt.bar(categorias, valores, color=['#1f77b4','#ff7f0e','#2ca02c'])
plt.title("Colores bien diferenciados")
plt.show()
```

## 4. Mostrar valores sin contexto (porcentajes sin total)

```python
porcentajes = [40, 35, 25]
etiquetas = ['A','B','C']

plt.bar(etiquetas, porcentajes, color='violet')
plt.title("Porcentajes sin información del total")
plt.show()
```

### Solución

```python
total = 200
valores = [0.4*total, 0.35*total, 0.25*total]

plt.bar(etiquetas, valores, color='violet')
plt.title(f"Distribución de {total} ventas por categoría")
plt.ylabel("Número de ventas")
plt.show()
```
