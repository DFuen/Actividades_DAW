## 3. Gráficos básicos con Matplotlib

### 3.1 Gráfico de líneas (plot)

```python
meses = ['Ene','Feb','Mar','Abr','May']
ventas = [120, 135, 142, 150, 170]
```

```python
plt.plot(meses, ventas, color='teal', marker='o', linestyle='--')
plt.title("Evolución de ventas mensuales")
plt.xlabel("Mes")
plt.ylabel("Ventas (€)")
plt.grid(True)
plt.show()
```

### 3.2 Gráfico de barras (bar)

```python
categorias = ['A','B','C','D']
valores = [120, 90, 160, 130]
```

```python
plt.bar(categorias, valores, color='skyblue')
plt.title("Ventas por categoría")
plt.xlabel("Categoría")
plt.ylabel("Unidades vendidas")
plt.show()
```

### 3.3 Gráfico de barras horizontales (barh)

```python
productos = ['Teléfonos','Portátiles','Tablets','Smartwatches']
ventas = [150, 200, 130, 90]
```

```python

plt.barh(productos, ventas, color='lightgreen')
plt.title("Ventas por tipo de producto")
plt.xlabel("Unidades vendidas")
plt.show()
```

### 3.4 Gráfico circular (pie)

```python
porcentajes = [40, 25, 20, 15]
categorias = ['A','B','C','D']
```

```python
plt.pie(porcentajes, labels=categorias, autopct='%1.1f%%', startangle=90)
plt.title("Distribución porcentual de ventas")
plt.show()
```

### 3.5 Diagrama de dispersión (scatter)

```python
import numpy as np
```

```python
x = np.random.rand(50)
y = x * 2 + np.random.randn(50)
```

```python
plt.scatter(x, y, color='orange')
plt.title("Relación entre dos variables")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.show()
```

### 3.6 Histograma (hist)

```python
datos = np.random.randn(500)
```

```python
plt.hist(datos, bins=20, color='violet', edgecolor='black')
plt.title("Distribución de datos aleatorios")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

## 4. Personalización del gráfico

### 4.1 Títulos, ejes y leyendas

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 20, 18, 25]
```

```python
plt.plot(x, y, label="Ventas 2025", color='navy', marker='s')
plt.title("Evolución de ventas", fontsize=14)
plt.xlabel("Meses", fontsize=12)
plt.ylabel("Ingresos (€)", fontsize=12)
plt.legend()
plt.grid(True)
plt.show()
```

### 4.2 Colores y estilos

```python
plt.plot(x, y, color='#FF5733', linestyle='-.', marker='o', linewidth=2)
plt.title("Ejemplo de personalización")
```

### 4.3 Anotaciones

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 20, 18, 22, 25]

plt.plot(x, y, marker='o', color='steelblue')
plt.title("Evolución de ventas con anotación")
plt.xlabel("Mes")
plt.ylabel("Ventas (miles €)")

# Anotación mejor colocada y con separación del punto
plt.annotate(
    'Pico máximo',               # texto de la anotación
    xy=(5, 25),                  # punto al que apunta la flecha
    xytext=(3.5, 27),            # posición del texto
    textcoords='data',           # coordenadas relativas a los ejes
    arrowprops=dict(facecolor='black', shrink=0.05, width=1, headwidth=8),
    fontsize=10,
    color='black'
)

plt.tight_layout()
plt.show()
```

## 5. Subgráficos y figuras múltiples

```python
import matplotlib.pyplot as plt

# Crear la figura con 1 fila y 2 columnas
fig, axs = plt.subplots(1, 2, figsize=(10, 4))

# Primer subgráfico (barras)
axs[0].bar(['A', 'B', 'C'], [5, 7, 3], color='lightcoral')
axs[0].set_title("Gráfico de barras")
axs[0].set_xlabel("Categorías")
axs[0].set_ylabel("Valores")

# Segundo subgráfico (líneas)
axs[1].plot([1, 2, 3], [3, 2, 5], color='steelblue', marker='o')
axs[1].set_title("Gráfico de líneas")
axs[1].set_xlabel("Eje X")
axs[1].set_ylabel("Eje Y")

# Ajustar espacios y mostrar
plt.tight_layout()
plt.show()
```

## 10. Propiedades y parámetros más utilizados en Matplotlib

### 1. Propiedades generales de las figuras y los ejes

```python
import matplotlib.pyplot as plt

# Crear una figura de tamaño personalizado
plt.figure(figsize=(7, 5))

# Dibujar un gráfico sencillo
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 4, 6]
plt.plot(x, y, label="Serie 1")

# Título y etiquetas de los ejes
plt.title("Ejemplo de propiedades generales", fontsize=14)
plt.xlabel("Eje X - Valores de entrada", fontsize=12)
plt.ylabel("Eje Y - Resultados", fontsize=12)

# Cuadrícula y leyenda
plt.grid(True)
plt.legend()

plt.show()
```

### 2. Parámetros visuales y de estilo

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [10, 12, 8, 14, 13]

plt.plot(
    x, y,
    color='teal',          # color de línea
    linestyle='--',        # tipo de línea
    linewidth=2,           # grosor
    marker='o',            # marcador
    markersize=8,          # tamaño del marcador
    alpha=0.8,             # transparencia
    label="Serie A"
)

plt.title("Parámetros visuales y de estilo")
plt.legend()
plt.show()
```

### 3. Parámetros específicos según el tipo de gráfico

#### Gráficos de líneas (plot)

```python
import matplotlib.pyplot as plt

meses = ['Ene', 'Feb', 'Mar', 'Abr', 'May']
ventas = [120, 140, 150, 130, 160]

plt.plot(meses, ventas, color='navy', linestyle='-', marker='s', label='Ventas 2025')
plt.title("Evolución de ventas (plot)")
plt.xlabel("Mes")
plt.ylabel("Unidades vendidas")
plt.legend()
plt.show()
```

#### Gráficos de barras (bar, barh)

```python
categorias = ['A', 'B', 'C', 'D']
valores = [50, 70, 30, 90]

plt.bar(categorias, valores, color='salmon', width=0.5, edgecolor='black')
plt.title("Comparativa de categorías (bar)")
plt.xlabel("Categorías")
plt.ylabel("Valores")
plt.show()
```

```python
productos = ['Teléfono', 'Tablet', 'PC', 'Smartwatch']
ventas = [150, 100, 200, 80]

plt.barh(productos, ventas, color='lightgreen')
plt.title("Ventas por producto (barh)")
plt.xlabel("Unidades vendidas")
plt.show()
```

#### Gráficos circulares (pie)

```python
porcentajes = [40, 25, 20, 15]
categorias = ['A', 'B', 'C', 'D']

plt.pie(
    porcentajes,
    labels=categorias,
    autopct='%1.1f%%',
    startangle=90,
    explode=(0, 0.1, 0, 0)
)
plt.title("Distribución porcentual (pie)")
plt.show()
```

#### Histogramas (hist)

```python
import numpy as np

datos = np.random.normal(100, 15, 200)
plt.hist(datos, bins=10, color='orchid', edgecolor='black', density=False)
plt.title("Distribución de datos (hist)")
plt.xlabel("Valor")
plt.ylabel("Frecuencia")
plt.show()
```

#### Diagramas de dispersión (scatter)

```python
import numpy as np

x = np.random.rand(50)
y = 2 * x + np.random.randn(50) * 0.1

plt.scatter(x, y, s=80, c='orange', alpha=0.7, marker='^')
plt.title("Diagrama de dispersión (scatter)")
plt.xlabel("Variable X")
plt.ylabel("Variable Y")
plt.show()
```

### 4. Propiedades de personalización avanzada

```python
import matplotlib.pyplot as plt

x = [1, 2, 3, 4, 5]
y = [3, 6, 8, 7, 10]

plt.plot(x, y, marker='o', color='royalblue')
plt.title("Propiedades de personalización avanzada")

# Líneas de referencia
plt.axhline(0, color='black', linewidth=1)
plt.axvline(3, color='gray', linestyle='--')

# Texto y anotaciones
plt.text(2, 8, "Tendencia positiva", color='darkgreen')
plt.annotate(
    "Máximo local", 
    xy=(5, 10), 
    xytext=(3.5, 10.5),
    arrowprops=dict(facecolor='black', arrowstyle='->')
)

# Límites de ejes
plt.xlim(0, 6)
plt.ylim(0, 12)

plt.grid(True)
plt.show()
```
