```python
#Cargo el archivo actividad_usuarios.csv
actividad_usuarios=[]
#Cargar los datos con numpy
datos = np.genfromtxt('actividad_usuarios.csv', delimiter=',', skip_header=1)
print("Tabla de actividad_usuarios sin cabeceras")
print(datos[1])#muestro el primer grupo de datos
```

```python
# Extraemos columnas

publicaciones = datos[:, 1]

print("EJEMPLO DE VISUALIZACIÓN DATOS")
print()
print(publicaciones[4999], "publicaciones")

likes_recibidos = datos[:, 2]

print(likes_recibidos[4999], "likes_recibidos",)

comentarios_recibidos = datos[:, 3]

print(comentarios_recibidos[4999], "comentarios_recibidos")

horas_uso = datos[:, 5]

print(horas_uso[4999], "horas_uso")
```

```python
# 3. VERIFICACIÓN CON PANDAS
df_verificacion = pd.read_csv("actividad_usuarios.csv")

# 4. COMPARACIÓN Y VALIDACIÓN (NumPy vs Pandas)
print("--- VERIFICACIÓN DE DATOS (Últimas 5 filas) ---")

# Comparamos las últimas 5 filas de NumPy contra las de Pandas para asegurar consistencia
comparacion = pd.DataFrame({
    'NumPy_Horas': horas_uso[-5:],
    'Pandas_Horas': df_verificacion['horas_uso'].tail(5).values
})

print(comparacion)

# Comprobación lógica: ¿Son exactamente iguales?
son_iguales = np.array_equal(horas_uso, df_verificacion['horas_uso'].values)
print(f"\n¿Coinciden NumPy y Pandas exactamente?: {son_iguales}")
```

```python
#2.
```

```python
# 1. MANEJO DE CEROS PARA LA TASA REAL
# Usamos np.where para evitar el error de división por cero (inf/nan)
# Estructura: np.where(condición, cálculo_si_es_true, valor_si_es_false)
tasa_real = np.where(horas_uso > 0, 
                     (likes_recibidos + comentarios_recibidos) / horas_uso, 
                     0)

print("--- VALIDACIÓN DE TASA REAL ---")
print(f"Registro 4999 - Tasa Real: {tasa_real[4999]:.4f}")
```

```python
#3.
```

```python
# CÁLCULO DE MEDIA
media_global = np.mean(tasa_real)
print(f"MEDIA GLOBAL DE INTERACCIONES: {media_global:.3f}")
```

```python
# CORRECCIÓN: Contamos cuántos True hay (cada True vale 1)
encima_media = np.sum(tasa_real > media_global) 
print(f"{encima_media} usuarios están por encima de la media.")
```

```python
#4.
```

```python
# --- CORRECCIÓN DE CORRELACIONES (Extracción [0,1] y Formato) ---

# 1. Correlación entre publicaciones y likes
# Extraemos el valor [0,1] para obtener el coeficiente, no la matriz completa
corrc_pub_likes = np.corrcoef(publicaciones, likes_recibidos)[0, 1]
print(f"CORRELACIÓN ENTRE PUBLICACIONES Y LIKES: {corrc_pub_likes:.4f}")

# 2. Correlación entre likes y comentarios
corrc_likes_com = np.corrcoef(likes_recibidos, comentarios_recibidos)[0, 1]
print(f"CORRELACIÓN ENTRE LIKES Y COMENTARIOS:   {corrc_likes_com:.4f}")

# 3. Correlación entre horas de uso y tasa real
# Usamos la tasa_real que calculamos previamente (la que ya no tiene Infs)
corrc_uso_tasa = np.corrcoef(horas_uso, tasa_real)[0, 1]
print(f"CORRELACIÓN ENTRE USO Y TASA REAL:       {corrc_uso_tasa:.4f}")
```

#### 5.
##### * Si, ya que tienen correlación muestra que estan cerca de 1 que del 0 (1 correlación y 0 no tienen)
##### * No, ya que aunque sea inversa no tiene relación ya que está mas cerca del valor 0
