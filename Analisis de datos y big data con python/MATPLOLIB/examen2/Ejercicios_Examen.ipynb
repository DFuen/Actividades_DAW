```python
tasa_real=(likes_recibidos+comentarios_recibidos)/horas_uso
print("EJEMPLO DE DATOS TASA REAL")
print(tasa_real[4999], "Tasa_Real_Interaccion")#muestro el calculo de la tasa real del ultimo grupo de datos
print()
#Saco los valores de la tasa de interacción del archivo
tasa_interaccion=np.genfromtxt('actividad_usuarios.csv',delimiter=',',skip_header=1,usecols=(6))
print("DIFERENCIAS RELEVANTES")
print(tasa_real[4999],"Tasa_Real_Interaccion","//",tasa_interaccion[4999], "Tasa_Interacción(archivo)")
#se puede apreciar que la tasa que da el archivo es mayor que la real.
```

```python
corrc_publicaciones_likes=np.corrcoef(publicaciones,likes_recibidos)
print("CORRELACIÓN ENTRE PUBLICACIONES Y LIKES RECIBIDOS")
print(corrc_publicaciones_likes)
corrc_likes_comentarios=np.corrcoef(likes_recibidos,comentarios_recibidos)
print("CORRELACIÓN ENTRE LIKES RECIBIDOS Y COMENTARIOS RECIBIDOS")
print(corrc_likes_comentarios)
corrc_uso_interaccionReal=np.corrcoef(horas_uso,tasa_real)
print("CORRELACIÓN ENTRE HORAS DE USO Y TASA REAL DE INTERACCIONES")
print(corrc_uso_interaccionReal)
```
