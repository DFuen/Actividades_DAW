#### Ejercicio 7

#### Partiendo de una matriz de 10 filas y 10 columnas con valores aleatorios enteros en el intervalo [0,100] realiza las operaciones necesarias para obtener una matriz de las mismas dimensiones donde: 
#### Todos los elementos igual a 10 se les suma 200.
#### Los elementos entre 11 y 50 tengan valor 100.
#### Los elementos mayores que 50 tengan valor 0.
#### Los elementos menores de 10 tengan el valor 1.


```python
# Creamos la matriz de 10x10
# El rango es [0, 101) porque el límite superior es exclusivo
matriz = np.random.randint(0, 101, (10, 10))
print("Matriz Original:\n", matriz)
```

```python
# Hacemos una copia para no modificar la original
resultado = matriz.copy()

# 1. Elementos igual a 10: sumar 200
resultado[resultado == 10] += 200

# 2. Elementos entre 11 y 50: valor 100
# Usamos & (AND) para combinar condiciones
resultado[(resultado >= 11) & (resultado <= 50)] = 100

# 3. Elementos mayores que 50: valor 0
resultado[resultado > 50] = 0

# 4. Elementos menores de 10: valor 1
resultado[resultado < 10] = 1

print("\nMatriz Transformada:\n", resultado)
```

#### A partir de la matriz resultante del ejercicio anterior, calcula lo siguiente:

#### Cuenta cuántas personas tienen IMC < 18.5 (bajo peso), 18.5–24.9 (normal), 25–29.9 (sobrepeso) y ≥30 (obesidad).

```python
# Extraemos la columna de IMC para trabajar más fácil
columna_imc = matriz_completa[:, 2]

# Bajo peso (IMC < 18.5)
bajo_peso = (columna_imc < 18.5).sum()

# Peso normal (18.5 <= IMC < 25)
# Usamos & para indicar "y esto también"
peso_normal = ((columna_imc >= 18.5) & (columna_imc < 25)).sum()

# Sobrepeso (IMC >= 25)
sobrepeso = (columna_imc >= 25).sum()

print(f"Bajo peso: {bajo_peso}")
print(f"Peso normal: {peso_normal}")
print(f"Sobrepeso: {sobrepeso}")
```
